```python
no_of_passengers = tit_df['survived'].count()
total_male = tit_df.loc[tit_df['sex']=='male']
total_male_alive = total_male.loc[total_male['alive']=='yes','alive'].count()
total_female = tit_df.loc[tit_df['sex']=='female']
total_female_alive = total_female.loc[total_female['alive']=='yes','alive'].count()
total_alive_passengers = tit_df.loc[tit_df['alive']=='yes','alive'].count()
not_alive_passengers = tit_df.loc[tit_df['alive']=='no','alive'].count()
print(no_of_passengers)
print(total_alive_passengers)
print(not_alive_passengers)
print(total_male_alive)
print(total_female_alive)
```

```python
# Pie Plot For Alive and not Alive Passengers

plt.pie([total_alive_passengers,not_alive_passengers],
        labels=['total_alive_passengers','not_alive_passengers'],
        autopct="%0.1f%%",
        explode=[0,0.02],
        colors = ['g','r'],
        shadow=True,
        wedgeprops={'linewidth':2,'edgecolor':'k'}
       )
plt.show()
```

```python
# Male and Female alive Percent

plt.pie([total_male_alive,total_female_alive,not_alive_passengers],
        labels=['total_male_alive','total_female_alive','not_alive_passengers'],
        autopct="%0.1f%%",
        explode=[0,0.02,0],
        colors = ['g','b','r'],
        shadow=True,
        wedgeprops={'linewidth':2,'edgecolor':'k'}
       )
plt.show()
```

```python
# Pie Chart for 1st class 2nd class and 3rd class alive 

f_class = tit_df.loc[tit_df['class']=='First']
f_class_alive = f_class.loc[f_class['alive']=='yes','alive'].count()

s_class = tit_df.loc[tit_df['class']=='Second']
s_class_alive = s_class.loc[s_class['alive']=='yes','alive'].count()

t_class = tit_df.loc[tit_df['class']=='Third']
t_class_alive = t_class.loc[t_class['alive']=='yes','alive'].count()

print(f_class_alive)
print(s_class_alive)
print(t_class_alive)
```

```python
plt.pie([f_class_alive,s_class_alive,t_class_alive,not_alive_passengers],
        labels=['f_class_alive','s_class_alive','t_class_alive','not_alive_passengers'],
        autopct="%0.1f%%",
        explode=[0,0,0,0.02],
        shadow=True,
        wedgeprops={'linewidth':2,'edgecolor':'k'}
       )
plt.show()
```

```python
# Pie Chart for alone and not alone passengers

alone_alive = tit_df.loc[tit_df['alone']==True]
alone_alive = alone_alive.loc[alone_alive['alive']=='yes','alive'].count()

not_alone_alive = tit_df.loc[tit_df['alone']==False]
not_alone_alive = not_alone_alive.loc[not_alone_alive['alive']=='yes','alive'].count()


plt.pie([alone_alive,not_alone_alive,not_alive_passengers],
        labels=['alone_alive','not alone_alive','not_alive_passengers'],
        autopct="%0.1f%%",
        explode=[0,0,0.02],
        shadow=True,
        wedgeprops={'linewidth':2,'edgecolor':'k'}
       )
plt.show()
```
